 PART 2: DATA CLEANING & FEATURE ENGINEERING
1 Fix Data Types

```python
df['date_issued']=pd.to_datetime(df['date_issued'])
df['date_last_payment']=pd.to_datetime(df['date_last_payment'])
```

```python
df['payment_gap_days']= (df['date_issued']-df['date_last_payment']).dt.days
```

```python
df['income-loan_ration'] = df['income']-df['disbursed_amount']
```

```python
df.drop(columns=['customer_id'],inplace=True)
```

```python
df.head(3)
```

 PART 3: MACHINE LEARNING PIPELINE

```python
X = df.drop('target',axis=1)
y = df['target']
```

Identify Columns

```python
num_features = X.select_dtypes(include=['Float64','int64']).columns
cat_features = X.select_dtypes(include=['object']).columns
```

3 Build Preprocessing Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
numeric_pipeline = Pipeline([('scaler',StandardScaler())])
categorical_pipeline = Pipeline([('encoder',OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer([('num',numeric_pipeline,num_features),('cat',categorical_pipeline,cat_features)])

```

```python
from sklearn.ensemble import RandomForestClassifier
model = Pipeline([('preprocessing' , preprocessor),('classifier',RandomForestClassifier(n_estimators=200,class_weight='balanced',random_state=42))])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)

model.fit(X_train,y_train)
```

6 Model Evaluation

```python
from sklearn.metrics import classification_report, accuracy_score,confusion_matrix

y_pred = model.predict(X_test)

print("Classifiaction report :\n",classification_report(y_test,y_pred))
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True,fmt='d')
plt.title("Confusion Matrix")
plt.show()
```

```python
print("accuracy_score =\n",accuracy_score(y_test,y_pred))
```
